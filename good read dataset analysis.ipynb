```python
# Import pandas library
import pandas as pd

# Read the first csv file and store it in a DataFrame object called df1
df1 = pd.read_csv(r"audible datasets\audible_cleaned.csv")

# Read the second csv file and store it in a DataFrame object called df2
df2 = pd.read_csv(r"books_1.Best_Books_Ever.csv")

# Use the pandas.merge function to join the two DataFrames on the bookId column, using the outer join method to keep all the rows from both DataFrames
merged_df = pd.merge(df1, df2, on='bookId', how='outer')

# Write the merged DataFrame to a new csv file called merged.csv
merged_df.to_csv(r"merged datasets\merged.csv", index=False)
```
